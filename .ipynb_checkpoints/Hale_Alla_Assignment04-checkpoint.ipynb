## Problem Statement

Ford GoBike (https://www.fordgobike.com/), the company running Bay Area Bikeshare, is trying to increase ridership.  To accomplish this, Ford GoBike wants to offer deals through the mobile app. What deals should be offered?

Currently, the company has three options: a flat price for a single one-way trip, a day pass that allows unlimited 30-minute rides for 24 hours and an annual membership. 


## Overall Recommendations

The detailed analysis supporting these recommendations follows.  From the available dataset, it is clear that ridership has declined in the last several months.  

In order to increase ridership, it is recommended to:

1. **Cater to subscribers\* with any offers.**  Overall, close to 90 % of all bike trips are taken by subscribers.  Even when total numbers drop, subscribers are still taking the majority of trips.  They are important to keep, and it is important to convert more customers to subcribers.  
\* Subscribers are those with annual or 30-day memberships; customers are those with daily or 3-day passes.

2. **Take advantage of peak and low demand times.** The majority of rides occur on weekdays, during rush hours (7-9 AM and 5-7 PM).  Ford GoBike should offer deals during low demand times to increase ridership at those times.  Additionally, Ford GoBike could increase rates during peak demand times to increase revenues.  Alternatively, planning bike maintenance at the low demand times would allow for continued service quality while maintaining the bike fleet.

3. **Offer holiday specials.**  Holidays, in particular, show low ridership rates.  Days like Christmas, Independence Day, and others that are typically non-working days correspond with low ridership rates.  On these days, it would be good to offer specials to encourage people to try out the service.

4. **Offer weekend-only subscriptions.** Customers ride on the weekends, but it seems that overall rides could increase if we offered a weekend-only subscription.  For those people who do use the bikes, but only on the weekends, the subscription rate could be significantly less than the full subscription rate, but higher than a day pass.  This would be tempting for riders who might want to subscribe for more frequent use, but not for the purposes of commuting.

5. **Add a pay per time option.** Most trips are less than 30 minutes in length.  Since the day pass allows 30-minute rides, there may be potential riders who choose to walk over paying for a 30-minute ride but using only 5 minutes of it.  Adding a pay per time option may entice more riders to hop on bikes for the very short trips.

## How has ridership changed over time?
Have the number of trips remained constant over the years?  Or have they increased or declined?  What about the number of subscribers? Have we gained or lost riders?

```python
# How has ridership changed over time?
! bq query --use_legacy_sql=FALSE --format=csv --max_rows=3000 \
'with subscriber_table as \
(SELECT EXTRACT(date from start_date) as day,\
COUNT(DISTINCT trip_id) as subscriber_trips \
FROM `bigquery-public-data.san_francisco.bikeshare_trips` \
WHERE subscriber_type = "Subscriber" \
GROUP BY day \
ORDER BY day), \
\
customer_table as \
(SELECT EXTRACT(date from start_date) as day, \
COUNT(DISTINCT trip_id) as customer_trips \
FROM `bigquery-public-data.san_francisco.bikeshare_trips` \
WHERE subscriber_type = "Customer" \
GROUP BY day \
ORDER BY day) \
\
SELECT EXTRACT(date from start_date) as day, \
st.subscriber_trips, \
ct.customer_trips, \
COUNT(DISTINCT trip_id) as total_trips \
FROM `bigquery-public-data.san_francisco.bikeshare_trips` bt \
LEFT JOIN subscriber_table st on st.day = EXTRACT(date from bt.start_date) \
LEFT JOIN customer_table ct on ct.day = EXTRACT(date from bt.start_date) \
GROUP BY 1,2,3 \
ORDER BY 1' > ridership.csv
```

```python
# Read in the ridership table.
ridership = pd.read_csv("ridership.csv")
ridership.head()
```

```python
# Plot ridership over time
fig, ax = plt.subplots(figsize=(20, 6))
ax = ridership.plot(ax = ax, x=ridership['day'], sort_columns=True)
plt.xlabel('Date')
plt.ylabel('Trips per Day')
plt.title('Daily Trips Over Time')
ax.legend()
```

After an initial start up period, the number of trips taken per day has stabilized into a repeatable pattern.  Weekdays have much higher demand as shown before.  Weekends and holidays (especially Christmas) show much lower demand time.  Offering deals during these holiday times could increase ridership.  

During the start up period most of the trips taken were taken by customers and not subscribers, there were fewer total trips.  It appears that users were learning to use the bike sharing service.  Now, most of the trips appear to be taken by subscribers.  

There is a week in 2016 that appears to have a large spike of customer trips.  It would be informative to take a look at what happened that week to understand why there was a sudden spike in customer trips, without a corresponding uptick in total trips.  

Additionally, there is a marked decrease in ridership in 2016 over 2015.  While 2015 had ~1200 trips per day, 2016 has closer to ~1000 trips per day.  In order to reverse this trend, it is necessary to understand what is actually happening.  Is there a new, competing service?  Or perhaps a study on biker safety came out that is deterring bikers.  Or, although unlikely, perhaps traffic suddenly got much better and there is reduced demand for bikes.  

```python
# How about the composition of the rider base?  
# What is the percentage of subscribers over time?
ridership['pct_subscriber'] = ridership['subscriber_trips']/ridership['total_trips']*100
ridership_2 = ridership.set_index(['day'])
ridership_3 = ridership_2[ridership_2.columns[3]]
ridership_3.head()
                        
#Plot
fig, ax = plt.subplots(figsize=(20, 6))
ax = ridership_3.plot()
plt.xlabel('Date')
plt.ylabel('Percent Subscribers')
plt.title('Percent Subscribers over Time')
```

The percentage of subscriber trips over time is remarkably steady after the start-up period, and hovers right around 87 %.  Interestingly, the percentage of subscriber trips dips on the weekends and holidays.  This indicates that subscribers are likely the ones commuting, and more casual customers may use the service at off peak times.  This supports the recommendation to offer deals at off peak times, since the casual customers are more likely to be swayed by these deals.  Furthermore, subscribers will not generate additional revenue if they increase the amount of off-peak rides they take.  

These charts are informative, but the weekends add lots of noise.  So, to get a clearer picture, aggregate by month.

```python
! bq query --use_legacy_sql=FALSE --format=csv --max_rows=3000 \
'with subscriber_table as \
(SELECT EXTRACT(date from start_date) as day,\
COUNT(DISTINCT trip_id) as subscriber_trips \
FROM `bigquery-public-data.san_francisco.bikeshare_trips` \
WHERE subscriber_type = "Subscriber" AND \
EXTRACT(dayofweek from start_date) in (2, 3, 4, 5, 6) \
GROUP BY day \
ORDER BY day), \
\
customer_table as \
(SELECT EXTRACT(date from start_date) as day, \
COUNT(DISTINCT trip_id) as customer_trips \
FROM `bigquery-public-data.san_francisco.bikeshare_trips` \
WHERE subscriber_type = "Customer" AND \
EXTRACT(dayofweek from start_date) in (2, 3, 4, 5, 6) \
GROUP BY day \
ORDER BY day) \
\
SELECT EXTRACT(date from start_date) as day, \
st.subscriber_trips, \
ct.customer_trips, \
COUNT(DISTINCT trip_id) as total_trips \
FROM `bigquery-public-data.san_francisco.bikeshare_trips` bt \
LEFT JOIN subscriber_table st on st.day = EXTRACT(date from bt.start_date) \
LEFT JOIN customer_table ct on ct.day = EXTRACT(date from bt.start_date) \
WHERE EXTRACT(dayofweek from start_date) in (2, 3, 4, 5, 6) \
GROUP BY 1,2,3 \
ORDER BY 1' > ridership_weekday.csv
```

```python
ridership_weekday = pd.read_csv("ridership_weekday.csv")
ridership.head()

#Plot
fig, ax = plt.subplots(figsize=(20, 6))
ax = ridership_weekday.plot(ax = ax, x=ridership_weekday['day'], sort_columns=True)
plt.xlabel('Date')
plt.ylabel('Trips per Day')
plt.title('Daily Trips Over Time, Weekdays Only')
ax.legend()


ridership_weekday['pct_subscriber'] = ridership_weekday['subscriber_trips']/ridership_weekday['total_trips']*100
ridership_w_2 = ridership_weekday.set_index(['day'])
ridership_w_3 = ridership_w_2[ridership_w_2.columns[3]]
ridership_w_3.head()
                        
#Plot
fig, ax = plt.subplots(figsize=(20, 6))
ax = ridership_w_3.plot()
plt.xlabel('Date')
plt.ylabel('Percent Subscribers')
plt.title('Percent Subscribers Over Time, Weekdays Only')
```

Excluding weekends cleans up the picture well.  You can still see all of the major, non-working holidays throughout the year.  Most notable of the holidays are Christmas Day and New Year's Day.

## Customers vs. Subscribers
Who is riding when?  How do we turn more customers into subscribers?  Are there other kinds of subscriptions we could offer?

```python
! bq query --use_legacy_sql=false --format=csv \
'SELECT \
CASE WHEN EXTRACT(DAYOFWEEK from start_date) IN (1,7) THEN "weekend" \
WHEN EXTRACT(DAYOFWEEK from start_date) IN (2,3,4,5,6) THEN "weekday" \
ELSE null END as daytype,\
CASE WHEN EXTRACT(hour from start_date) IN (7, 8, 9) THEN "morning"\
WHEN EXTRACT(hour from start_date) IN (17, 18, 19) THEN "evening"\
ELSE null END as rush,\
COUNT(distinct EXTRACT(date from start_date)) as days,\
COUNT(distinct trip_id) as trips,\
COUNT(distinct trip_id)/COUNT(distinct EXTRACT(date from start_date)) as trips_per_day,\
subscriber_type \
FROM `bigquery-public-data.san_francisco.bikeshare_trips`\
GROUP BY daytype, rush, subscriber_type'  > trips_cat.csv  
```

```python
trips_cat = pd.read_csv("trips_cat.csv")
trips_cat
```

```python
sns.factorplot(x="rush", y="trips_per_day", hue="subscriber_type", col="daytype", data=trips_cat, kind='bar');
```

Most trips are taken in the rush hours (7-9 AM and 5-7 PM) on weekdays.  Comparing the same time periods on weekends shows that the number of trips is not nearly as many as on the weekdays.  Also, the number of subscribers taking rush hour trips is much larger than the number of customers.  To increase rush hour ridership, we need to increase the number of subscribers.  

However, to increase the overall ridership, it might be more useful to increase ridership during non-peak times.  We should offer deals to entice more one-off customer trips during the non-peak times.  Hopefully, when the customers understand how great the service is, they may use it more frequently on weekends and other non-peak times.  They might even start using the service for their commute.

Only 7539 trips have a duration of < 2 minutes, and only 72 trips have a duration of exactly 1 minute.  It's okay to not have removed these trips since they don't comprise a large proportion of the dataset.

```python
! bq query --use_legacy_sql=false --format=csv --max_rows=30000000 \
'SELECT duration_sec/60 as duration_min \
FROM `bigquery-public-data.san_francisco.bikeshare_trips` \
WHERE EXTRACT(dayofweek from start_date) IN (2,3,4,5,6) AND \
EXTRACT(hour from start_date) IN (7,8,9, 17,18,19) \
ORDER BY duration_min desc'\ > duration_rush.csv
```

```python
duration_rush = pd.read_csv("duration_rush.csv")
duration_rush.head()
```

```python
plt.hist(duration_rush['duration_min'], bins = 30, log=True)
plt.xscale('log')
plt.xlabel('Trip Duration, minutes')
plt.ylabel('Trip Count')
plt.title('Trip Duration')
```

```python
duration_rush.describe()
```

Most trips are pretty short.  We will examine just commuter trips, those that are under 1 hour. 

```python
! bq query --use_legacy_sql=false --format=csv --max_rows=30000000 \
'SELECT duration_sec/60 as duration_min \
FROM `bigquery-public-data.san_francisco.bikeshare_trips` \
WHERE EXTRACT(dayofweek from start_date) IN (2,3,4,5,6) AND \
EXTRACT(hour from start_date) IN (7,8,9, 17,18,19) AND \
duration_sec/60 < 60 \
ORDER BY duration_min desc'\ > duration_rush_short.csv
```

```python
duration_rush_short = pd.read_csv("duration_rush_short.csv")
duration_rush_short.head()
```

```python
duration_rush_short.describe()
```

Perhaps rides could be charged by 5 minute increments instead of a flat rate for a 30 minute ride.  That may entice riders to take very short rides instead of walking or taking an Uber.
